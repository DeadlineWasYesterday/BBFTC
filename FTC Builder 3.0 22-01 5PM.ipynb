```python
##Load metadata from file

import ast

d = {}
nary = {}
with open("metadata_trial4.txt", 'r') as f:
    for line in f:
        d = (ast.literal_eval(line))
        nary.update({d['id'] : d})
        
nary_all = nary
```

```python
#empty children class for nochilds

for key, value in nary.items():
    if value.get('children') == None:
        value['children'] = []
```

```python
#remove anything that has no aspect or is not a biological_process or molecular_function

c = 0
keys = []
for key, value in nary.items():
    if value.get('aspect') == None:
            keys.append(key)
            c+=1
    elif value['aspect'] != 'biological_process' and value['aspect'] != 'molecular_function':
        keys.append(key)
        c+=1
        
print(c)

for key, value in nary.items():
    for i in value['children']:
        if i['id'] in keys:
            c+=1
        elif value.get('babies') == None:
            value['babies'] = []
            value['babies'].append(i)
        else:
            value['babies'].append(i)
for i in keys:
    del nary[i]
```

```python
#Filter terms that have babies

filnary = {}
kilnary = {}
nochild = {}

for key, value in nary.items():
    if value.get('babies') != None:
        filnary = ({'id' : value['id'], 'babies' : value['babies'], 'name' : value['name'], 'aspect' : value['aspect']})
        kilnary.update({key : filnary})
    else:
        filnary = ({'id' : value['id'],})
        nochild.update({key : filnary})
```

```python
#get parents

for key, value in kilnary.items():
    for item in value['babies']:
        if nary[item['id']].get('parent') == None:
            nary[item['id']]['parent'] = []
            nary[item['id']]['parent'].append({'id' : key, 'relation' : item['relation']})
        else:
            nary[item['id']]['parent'].append({'id' : key, 'relation' : item['relation']})
            
for key, value in nary.items():
    if value.get('parent') == None:
        value['parent'] = []
```

```python
#empty children class for nochilds

for key, value in nary.items():
    if value.get('babies') == None:
        value['babies'] = []
```

```python
#get rid of the two motherfuckers

for k,v in nary.items():
    for item in v['parent']:
        for item2 in v['babies']:
            if item == item2:
                del(v['parent'][v['parent'].index(item)])
                del(v['babies'][v['babies'].index(item)])
```

```python
#remove childless duplicates

names = []
duplicates = {}

for k,v in nary.items():
    if v['name'] not in names:
        names.append(v['name'])
        
        
    for n in names:
        if n == v['name']:
            if duplicates.get(v['name']) == None:
                duplicates[v['name']] = []
                duplicates[v['name']].append(k)
            else:
                duplicates[v['name']].append(k)
```

```python
keys_for_deleting = []        
for k,v in duplicates.items():
    if len(v) <= 1:
        keys_for_deleting.append(k)
        
for k in keys_for_deleting:
    del duplicates[k]
```

```python
#define core classes

with onto:
    class therapeutic_agent(Thing):
        namespace = onto
    class biological_process_main(Thing):
        namespace = onto
    class molecular_function_main(Thing):
        namespace = onto
    class drug(Thing):
        namespace = onto
```

```python
def recursive_classmaker_down(term):
    for item in nary[term]['babies']:
        if item['relation'] == 'is_a':
            with onto:
                nc = types.new_class(item['id'], (getattr(onto, term),))
            recursive_classmaker_down(item['id'])
    return 0
```

```python
#Using classmaker to make FTC classes

for key, value in nary.items():
    if value['parent'] == [] and value['aspect'] == 'biological_process':
        with onto:
            nc = types.new_class(key, (biological_process_main,))
        for item in value['babies']:
            if item['relation'] == 'is_a':
                with onto:
                    nc = types.new_class(item['id'], (getattr(onto, key),))
                recursive_classmaker_down(item['id'])
    if value['parent'] == [] and value['aspect'] == 'molecular_function':
        with onto:
            nc = types.new_class(key, (molecular_function_main,))
        for item in value['babies']:
            if item['relation'] == 'is_a':                
                with onto:
                    nc = types.new_class(item['id'], (getattr(onto, key),))
                recursive_classmaker_down(item['id'])
        
```
